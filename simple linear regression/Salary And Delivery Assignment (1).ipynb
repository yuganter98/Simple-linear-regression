```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
```

```python
salary=pd.read_csv('D:\\Assignment Excelr\\simple linear regression\\Salary_Data.csv')
```

```python
salary
```

```python
salary.head()
```

```python
salary.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (5,5))
sns.scatterplot(x = 'Salary', y='YearsExperience',data=salary,palette='Sets1')
```

```python
salary.corr()
```

```python
import seaborn as sns
sns.distplot(salary['Salary'])
```

```python
sns.distplot(salary['YearsExperience'])
```

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols("YearsExperience~Salary",data = salary).fit()
```

```python
sns.regplot(x='Salary',y='YearsExperience',data=salary);
```

```python
model.params
```

```python
model.fittedvalues
```

```python
model.resid
```

```python
salary['Predicted']=model.fittedvalues
```

```python
salary['Predict']=-2.383161 + (0.000101 * salary['Salary'])
```

```python
salary
```

```python
salary['error']=model.resid
```

```python
salary['error']=salary['YearsExperience']- salary['Predicted']
```

```python
salary
```

```python
model.rsquared
```

```python
model.pvalues
```

```python
salary['Error_sq']=(salary['error'])**2
salary
```

```python
salary['Error_sq'].sum()
```

```python
newdata=pd.Series([200,300])
newdata
```

```python
salary1=pd.DataFrame(newdata,columns=['Salary'])
salary1
```

```python
model.predict(salary1)
```

```python
salary1['predicted']=model.predict(salary1)
salary1
```
